```python

## diffusion

betas = gd.get_named_beta_schedule(model_args.noise_schedule, model_args.diffusion_steps)
alpha1 = 1-betas
alpha_bar1 = np.cumprod(alpha1,axis=0)

if model_args.use_kl:
    loss_type = gd.LossType.RESCALED_KL
elif model_args.rescale_learned_sigmas:
    loss_type = gd.LossType.RESCALED_MSE
else:
    loss_type = gd.LossType.MSE
if not model_args.timestep_respacing:
    timestep_respacing = [model_args.diffusion_steps]
diffusion =  SpacedDiffusion(
    use_timesteps=space_timesteps(model_args.diffusion_steps, model_args.timestep_respacing),
    betas=betas,
    model_mean_type=(
        gd.ModelMeanType.EPSILON if not model_args.predict_xstart else gd.ModelMeanType.START_X
    ),
    model_var_type=(
        (
            gd.ModelVarType.FIXED_LARGE
            if not model_args.noise_schedule
            else gd.ModelVarType.FIXED_SMALL
        )
        if not model_args.learn_sigma
        else gd.ModelVarType.LEARNED_RANGE
    ),
    loss_type= loss_type,
    rescale_timesteps=model_args.rescale_timesteps,
)
```

```python
cls_args = args_to_dict(args, classifier_defaults().keys())
cls_args = edict(cls_args)
```

```python
if cls_args.image_size == 512:
    channel_mult = (0.5, 1, 1, 2, 2, 4, 4)
elif cls_args.image_size == 256:
    channel_mult = (1, 1, 2, 2, 4, 4)
elif cls_args.image_size == 128:
    channel_mult = (1, 1, 2, 3, 4)
elif cls_args.image_size == 64:
    channel_mult = (1, 2, 3, 4)
else:
    raise ValueError(f"unsupported image size: {image_size}")

attention_ds = []
for res in cls_args.classifier_attention_resolutions.split(","):
    attention_ds.append(cls_args.image_size // int(res))

classifier =  EncoderUNetModel(
    image_size= cls_args.image_size,
    in_channels=3,
    model_channels=cls_args.classifier_width,
    out_channels=1000,
    num_res_blocks=cls_args.classifier_depth,
    attention_resolutions=tuple(attention_ds),
    channel_mult=channel_mult,
    use_fp16=cls_args.classifier_use_fp16,
    num_head_channels=64,
    use_scale_shift_norm=cls_args.classifier_use_scale_shift_norm,
    resblock_updown=cls_args.classifier_resblock_updown,
    pool=cls_args.classifier_pool,
)
classifier.load_state_dict(
    dist_util.load_state_dict(args.classifier_path, map_location="cpu")
)
classifier.to(dist_util.dev())
if args.classifier_use_fp16:
    classifier.convert_to_fp16()
classifier.eval()
,
```

```python
NUM_CLASSES = 1000

def cond_fn(x, t, y=None):
    assert y is not None
    with th.enable_grad():
        x_in = x.detach().requires_grad_(True)
        logits = classifier(x_in, t)
        log_probs = F.log_softmax(logits, dim=-1)
        selected = log_probs[range(len(logits)), y.view(-1)]
        return th.autograd.grad(selected.sum(), x_in)[0] * args.classifier_scale

def model_fn(x, t, y=None):
    assert y is not None
    return model(x, t, y if args.class_cond else None)
```

```python
steps_gold = list(space_timesteps(1000,[250]))
steps_gold.sort()
```

```python
# some pre-define parameters
# steps = np.arange(250) *4 
betas = betas
alphas = 1.0- betas
alphas_bar = np.cumprod(alphas,axis=0)

#  respacing 之后的
new_betas = betas[steps_gold]
new_alphas = alphas[steps_gold]
new_alphas_bar = alphas_bar[steps_gold]

new_alphas_bar_pre = np.append(1.0,new_alphas_bar[:-1])  ## 为啥呢
new_alphas_bar_next = np.append(new_alphas_bar[1:],0.0) ## 为啥呢
new_sqrt_alphas_bar = np.sqrt(new_alphas_bar)
new_sqrt_one_minus_alphas_bar = np.sqrt(1-new_alphas_bar)

# ddpm 求解出来的方差 即 δ**2 表达式
post_variance = new_betas*(1.0-new_alphas_bar_pre)/(1.0-new_alphas_bar) 
posterior_log_variance_clipped = np.log(np.append(post_variance[1], post_variance[1:]))

# 求xstart 的参数， 前向递推公式可以求解出来 xt = sqrt(at_bar)*x0 + sqrt(1-at_bar) *noise
sqrt_recip_alphas_cumprod = np.sqrt(1.0 / new_alphas_bar)
sqrt_recipm1_alphas_cumprod = np.sqrt(1.0 / new_alphas_bar - 1)

# 再次求 均值用的 , ddpm 推理出来的均值的 u = f(x0,xt)  两项的系数
posterior_mean_coef1 = new_betas * np.sqrt(new_alphas_bar_pre) / (1.0 - new_alphas_bar)   # x0 的系数
posterior_mean_coef2 = (1.0 - new_alphas_bar_pre)* np.sqrt(new_alphas)/ (1.0 - new_alphas_bar)  # xt 的系数

# 直接用提供的呢
coeff_xt = 1.0/np.sqrt(new_alphas)
coeff_noise = -1.0*coeff_xt * (1-new_alphas)/np.sqrt(1-new_alphas_bar)
```
